### Parameters

```python
J = 50 # Numbers of the orders
```

Creating the amount of vegtables, fruits and others from poission destribution:

```python
lamb = [7.5, 5.2, 20] # Lambdas for the poission destribution
m = {1:[],2:[],3:[]} # 1: veggies, 2: fruits, 3: other
for i in m.keys(): # For each type of order
    for j in range(J): # For each order
        m[i].append(np.random.poisson(lamb[i-1]))
m_df = pd.DataFrame(m) # More easy to work with pandas df
m_df
```

Creating the velocity matrix for each order:

```python
v = {"veg":[],"fru":[], "oth":[]} # A dataframe of collector's velocity
for i in v.keys(): # For each type of order
    for j in range(J): # For each order
        if i == "veg":
            v[i].append((25/(25+25*m[1][j]))**1.2)
        elif i == "fru":
            v[i].append((25/(25+30*m[2][j]))**1.1)
        else:
            v[i].append(25/(25+10*m[3][j]))
v_df = pd.DataFrame(v) # More easy to work with pandas
v_df
```

Definig the length of each shelf:

```python
len_shelf = 25
len_pass = len_shelf*2
```

Definig the time metrix that takes to do each type of order, by using velpcity matrix:

```python
t = {"veg":[],"fru":[], "oth":[]} # A dataframe of collector's times
for i in t.keys(): # For each type of order
    for j in range(J): # For each order
        if i == "veg" or i == "fru": # Because veggies and fruits are in the same passage
            t[i].append(len_shelf/v_df[i][j])
        else: # Because others are in different passege
            t[i].append(len_pass/v_df[i][j])
t_df = pd.DataFrame(t)
t_df['Sum'] = t_df.sum(axis=1) # We will use it later and we will fix it to the real amount of time that will take to do the order
t_df       
```

### Question no. 1

For more info, look at the docs:

```python
def internal_arrangement(t_df, len_pass): # This function recives a time matrix, and we will use it for q2
    c = [0 for i in range(J)] # Definig end time matrix
    schedule = [0 for i in range(J)] # Defining the schedule for each order
    # Defining clock of the simulation
    t_now = 0 
    t_prev = 0
    for j in range(J): # For each order
        if m_df[1][j] > 0 or m_df[2][j] > 0: # If there are fruits or veggies to collect
            schedule[j] += t_now # Schedule the order for now
            t_prev = t_now 
            #if m_df[1][j] > 0: # If there are veggies, collect them
            t_now += t_df["veg"][j]
            #if m_df[2][j] > 0: # if ther are fruits, collect them
            t_now += t_df["fru"][j]
            if m_df[3][j] > 0: # If there are others, return back and collect them
                t_now += len_pass + t_df["oth"][j]
            t_df["Sum"][j] = t_now - t_prev # Fixing the sum collumn
            c[j] = t_now # Defining the end time of the order after all the collections
        else: # There are only others to collect
            t_now += t_df["oth"][j]
            c[j] = t_now # Defining the end time of the order after all the collections
    return c, schedule
c,schedule = internal_arrangement(t_df, len_pass)
max(c) # The answer to the question
```

Defining the spt algoritem function, we will use it also at question no. 2:

```python
def spt_algo(t_df):
    t_now = 0
    #t_prev = t_now
    t_copy = t_df["Sum"].copy() # Creating a copy of the fixed sum column, because we don't want to lose this info
    schedule = [0 for i in range(J)] # Defining the schedule for each order
    c = [0 for i in range(J)] # Definig end time matrix, only for proving a point
    for i in range(J): # For each order
        #t_prev = t_now
        index = t_copy.idxmin() # Pull the id of the minimal processing time
        schedule[index] += t_now # Scheduling the shortest order for current time
        t_now += t_copy[index] # Moving the sim clock to the end of the order
        c[i] = t_now # Definig the end time of the order
        w[i] = c[i] - t_df["Sum"][index] # Calculating the waiting time: end time - processing time
        print("c[i]: ",c[i])
        print("t_df['Sum'][index]: ", t_df["Sum"][index])
        t_copy[index] = None # "Deleting" the minimal processing time, so we can find the next minimal time
    return schedule, c, w
sche_spt,c_spt, w = spt_algo(t_df)
sche_spt
```

prep for Q//c_max - for more info go to doc:

calc the ratio of the task index (K) to the working rate of the gatherer (s_j: s1 = 1, s2 = 1.15): K/s_j

```python
K_S1 = [i for i in range(1, J+1)]

K_S2 = [0 for i in range(1, J+1)]
for i in range(1, 51):
    K_S2[i-1] = i/1.15
```

As we do in Q//sigma(C_i): take the task with the longest time and match it to the min ratio of both gatherers.

* The heuristics will be to place the tasks from the beginning and not the end (as we do in P//c_max). 

* Note the longest task will be the same for the both gatherers since the ratio is permanent

```python
#initializing schedules: 1 for the regular, 2 for the faster one
sch_1_Q3 = []
sch_2_Q3 = []

#initializing time counters
time_1 = 0
time_2 = 0

# to compare with Q4 - initializing dictionaries for waiting time to each task in specific gather. 
waiting_time_1 = {}
waiting_time_2 = {}

#copy for processing times
t_df_copy = t_df.copy()


for i in range(J):
    task_idx = t_df_copy['Sum'].idxmax() # task index with max time
    task_time_1 = t_df_copy['Sum'][task_idx] # time processing for the regular
    task_time_2 = t2_df['Sum'][task_idx] ## time processing for the fast gather

    t_df_copy['Sum'][task_idx] = None # for not using it again
    
    # taking min ratio
    min_K_S1 = min(K_S1)
    min_K_S2 = min(K_S2)
    
    min_tot = 0
    
    #case the min ratio is for the regular gather
    if min_K_S1 < min_K_S2:
        waiting_time_1[task_idx] = time_1 # waiting time until now, not including the process time for this task
        min_tot = min_K_S1
        idx = K_S1.index(min_tot)
        K_S1.pop(idx)
            
        sch_1_Q3.append(task_idx) # add the task to regular gather schedule
            
        time_1 += task_time_1 # update time
    
    #case the min ratio is for the fast gather    
    else:
        waiting_time_2[task_idx] = time_2 # waiting time until now, not including the process time for this task
        min_tot = min_K_S2
        idx = K_S2.index(min_tot)
        K_S2.pop(idx)
            
        sch_2_Q3.append(task_idx) # add the task to faster gather schedule
            
        time_2 += task_time_2 # update time


time_finish = max(time_2, time_1) # C_MAX

print('time_finish: ',time_finish)
print('sch for regular gather: ', sch_1_Q3)
print('sch for fast gather: ', sch_2_Q3)
```

As we do in Q//sigma(C_i): take the task with the longest time and match it to the min ratio of both gatherers.

* Note the longest task will be the same for the both gatherers since the ratio is permanent

```python
#initializing schedules: 1 for the regular, 2 for the faster one
sch_1_Q4 = []
sch_2_Q4 = []

#initializing time counters
time_1_Q4 = 0
time_2_Q4 = 0

# initializing dictionaries for waiting time to each task in specific gather. 
waiting_time_1_Q4 = {}
waiting_time_2_Q4 = {}

#copy for processing times
t_Q4_copy = t_df.copy()

### same algo as Q3 - no time clculations
for i in range(J):
    task_idx = t_Q4_copy['Sum'].idxmax() # task index with max time
    task_time_1 = t_Q4_copy['Sum'][task_idx] # time processing for the regular
    task_time_2 = t2_df['Sum'][task_idx] ## time processing for the fast gather
    
    t_Q4_copy['Sum'][task_idx] = None # for not using it again
    
    # taking min ratio
    min_K_S1 = min(K_S1_Q4)
    min_K_S2 = min(K_S2_Q4)
    
    min_tot = 0
    
    #case the min ratio is for the regular gather
    if min_K_S1 < min_K_S2:
        min_tot = min_K_S1
        idx = K_S1_Q4.index(min_tot)
        K_S1_Q4.pop(idx)
            
        sch_1_Q4.append([task_idx, task_time_1]) # add the task to regular gather schedule
    
    #case the min ratio is for the fast gather    
    else:
        min_tot = min_K_S2
        idx = K_S2_Q4.index(min_tot)
        K_S2_Q4.pop(idx)
            
        sch_2_Q4.append([task_idx,task_time_2]) # add the task to faster gather schedule
###

### New - Assign tasks from end to beginning
t1=0 # time (regular gather) counter
agg_time_1 = [0] # Auxiliary variable in order to calculate waiting time 
sch_1_rev = [] # Auxiliary variable to reverse scheduale
sch_1_final = []

for i in range(len(sch_1_Q4)):
    sch_1_rev.append(sch_1_Q4[-1])
    t1 += sch_1_Q4[-1][1]  # accumulation
    sch_1_Q4.pop(-1)
    
    agg_time_1.append(t1) # list of waiting time - no task index
    
    waiting_time_1_Q4[sch_1_rev[i][0]] = agg_time_1[i] # dict of waiting time- k= task, v= waiting time (reg gather)
    
    sch_1_final.append(sch_1_rev[i][0])


t2=0 # time (fast gather) counter
agg_time_2 = [0] # Auxiliary variable in order to calculate waiting time
sch_2_rev = [] # Auxiliary variable to reverse scheduale
sch_2_final = []

for i in range(len(sch_2_Q4)):
    sch_2_rev.append(sch_2_Q4[-1]) 
    t2 += sch_2_Q4[-1][1] # accumulation
    sch_2_Q4.pop(-1)
    
    agg_time_2.append(t2) # list of waiting time - no task index
    
    waiting_time_2_Q4[sch_2_rev[i][0]] = agg_time_2[i] # dict of waiting time- k= task, v= waiting time (fast gather)
    
    sch_2_final.append(sch_2_rev[i][0])


time_finish_Q4 = max(agg_time_1[-1], agg_time_2[-1]) # C_MAX

print('time_finish Q4: ', time_finish_Q4)
print('sch for regular gather: ', sch_1_final)
print('sch for fast gather: ', sch_2_final)
```

Depends on the assumptions and the work division between the workers -  C_max cannot be minimized. We make sure that each worker handles an order only if necessary, and calculate the finish time accordingly.

```python
#initializing schedules: 1 for the regular, 2 for the faster one
sch_1_Q5 = {i : None for i in range(50)}
sch_2_Q5 = {i : None for i in range(50)}

def internal_arrangement_Q5(t_df):
    # Defining clock of the simulation
    t1_now = 0
    t2_now = 0 
    c1 = [0 for i in range(J)] # Definig end time matrix - for worker 1
    c2 = [0 for i in range(J)] # Definig end time matrix - for worker 2
    c_total = [0 for i in range(J)] # Definig end time matrix - for each order - Max(c1_j,c2_j)
    schedule1 = [0 for i in range(J)] # Defining the schedule for each order - for worker 1 (the start times)
    schedule2 = [0 for i in range(J)] # Defining the schedule for each order - for worker 2 (the start times)
    w_total = [0 for i in range(J)] # Defining  waiting time for each order

    for i in range(J): # For each order 
        if m_df[1][i] > 0 or m_df[2][i] > 0: # If there are fruits or veggies to collect - for worker 2
            schedule1[i] += t1_now # Schedule the order for now
            t1_now += t5_df["Sum"][i]
            c1[i] = t1_now # Defining the end time of the order after all the collections
            sch_2_Q5[i]= i
        if m_df[3][i] > 0: # if there are others to collect - for worker 1
            schedule2[i] += t2_now # Schedule the order for now
            t2_now += t5_df["oth"][i]
            c2[i] = t2_now # Defining the end time of the order after all the collections
            sch_1_Q5[i]= i


    for i in range(J):
        c_total[i] = max(c1[i], c2[i])
        if ((schedule1[i] <= c2[i]) and (c1[i] >= schedule2[i])) or ((schedule2[i] <= c1[i]) and (c2[i] >= schedule1[i])):
            w_total[i] = min(schedule1[i], schedule2[i])
        else:
            processing_time = t5_df["oth"][i] + t5_df["Sum"][i]
            w_total[i] = c_total[i] - processing_time
    
    c_max_Q5 = max(c_total)

    return c1, c2, schedule1, schedule2, sch_1_Q5, sch_2_Q5, c_max_Q5, w_total
 
c1, c2, schedule1, schedule2, sch_1_Q5, sch_2_Q5, c_max_Q5, w_total_Q5 = internal_arrangement_Q5(t5_df)
```

### Question no. 6

```python
# Setting exclusive orders for each worker
only_worker1 = {} # Only worker 1 is responsible for the order - only "oth" 
only_worker2 = {} # Only worker 2 is responsible for the order - "veg", "fru"

for i in range(J): # For each order:
    if m_df[1][j] == 0 and m_df[2][j] == 0 and m_df[3][j] > 0: # If there are only "oth" in the order
        only_worker1[j] = t5_df["oth"][j]
    if (m_df[1][j] > 0 or m_df[2][j] > 0) and m_df[3][j] == 0: # If there are not "oth" in the order
        only_worker1[j] = t5_df["Sum"][j]
# Sort the orders from the shortest to the longest - for spt algo    
dict(sorted(only_worker1.items(), key=lambda item: item[1]))
dict(sorted(only_worker2.items(), key=lambda item: item[1]))

#initializing schedules: 1 for the regular, 2 for the faster one
sch_1_Q6 = {i : None for i in range(50)}
sch_2_Q6 = {i : None for i in range(50)}

def get_key(dic, val):
    for key, value in dic.items():
         if val == value:
             return key

def spt_algo_Q6(t_df):
    
    t1_now = 0 # Start Time of worker 1
    t2_now = 0 # Start Time of worker 1 
    t_oth_copy = t5_df["oth"].copy() # Creating a copy of  'Sum' column - for worker 1
    t_sum_copy = t5_df["Sum"].copy() # Creating a copy of 'oth' column - for worker 2
    c1 = [0 for i in range(J)] # Definig end time matrix - for worker 1
    c2 = [0 for i in range(J)] # Definig end time matrix - for worker 2
    c_total = [0 for i in range(J)] # Definig end time matrix - for each order - Max(c1_j,c2_j)
    schedule1 = [0 for i in range(J)] # Defining the schedule for each order - for worker 1 (the start times)
    schedule2 = [0 for i in range(J)] # Defining the schedule for each order - for worker 2 (the start times)
    w_total = [0 for i in range(J)] # Defining  waiting time for each order

    for i in range(J): # For each order
        index = t_oth_copy.idxmin() # Pull the id of the minimal processing time - for worker 1 
        schedule1[index] += t1_now # Scheduling the shortest order for current time
        t1_now += t_oth_copy[index] # Moving the sim clock to the end of the order
        c1[index] = t1_now 
        t_oth_copy[index] = None # "Deleting" the minimal processing time, so we can find the next minimal time
        sch_1_Q6[i]= index

        # If the order is also relevant for the worker 2
        if not i in only_worker1: 
            sch_2_Q6[i] = index

    for i in range (len(only_worker2)): # For each exclusive order of worker 2
        if only_worker1: # If there is an exclusive task of worker 1
            curr_key, curr_val = list(only_worker1.items())[0] # Worker1's shortest order
            only_worker1.pop(curr_key) # Pop the minimal processing time, so we can find the next minimal time
            rel_key = get_key(sch_1_Q6, curr_val) # Get the serial number of the current order by the schedule of worker 1
        else:
            rel_key = get_key(sch_2_Q6, None) # Get a serial number of "free time slot" of worker 2
        sch_2_Q6[rel_key] = list(only_worker2.keys())[i] # Insert the exclusive order of worker 2

    # Calculating times of worker 2
    for i in range(J):
        index = sch_2_Q6[i] # The shortest current order of worker 2 (sorted dictionary by keys)
        schedule2[index] += t2_now # Scheduling the shortest order for current time
        t2_now += t_sum_copy[index] # Moving the sim clock to the end of the order
        c2[index] = t2_now # Definig the end time of the order
    
    # Calculation of waiting times
    for i in range(J):
        c_total[i] = max(c1[i], c2[i])
        if ((schedule1[i] <= c2[i]) and (c1[i] >= schedule2[i])) or ((schedule2[i] <= c1[i]) and (c2[i] >= schedule1[i])):
            w_total[i] = min(schedule1[i], schedule2[i])
        else:
            processing_time = t5_df["oth"][i] + t5_df["Sum"][i]
            w_total[i] = c_total[i] - processing_time
    
    c_max_Q6 = max(c_total)

    return sch_1_Q6, schedule1, c1,  sch_2_Q6, schedule2, c2, c_max_Q6, w_total 
    
sch_1_Q6, schedule1_Q6, c1, sch_2_Q6, schedule2_Q6, c2_Q6, c_max_Q6, w_total_Q6 = spt_algo_Q6(t5_df)
```
